#### 프로젝트: 붕어빵 자동 판매 시스템


당신은 붕어빵 가게의 주인입니다. 

손님들이 붕어빵을 주문하면 이를 자동으로 처리하고, 재고를 관리하는 시스템을 만들고자 합니다. 

또한, 관리자 모드를 통해 재고를 추가할 수 있는 기능을 구현해야 합니다.


__[문제 1: 붕어빵 재고 확인]__

붕어빵 가게의 현재 재고를 확인하는 기능을 구현하세요.

요구사항:
- 현재 붕어빵의 재고를 딕셔너리로 관리합니다.
- 딕셔너리에는 붕어빵의 종류와 각각의 재고 개수가 저장됩니다.
- 현재 재고를 화면에 출력하세요.

붕어빵 재고
- 팥붕어빵 : 10개
- 슈크림붕어빵 : 8개
- 초코붕어빵 : 5개

```python
붕어빵_재고 = dict(팥=10, 슈크림=8, 초코=5)
붕어빵_관리 = 붕어빵_재고.items()

for a, b in 붕어빵_관리:
    print(f'{a}붕어빵: {b}개')
```

__[문제 2: 붕어빵 주문 받기]__

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
붕어빵_선택 = input('붕어빵 종류를 선택하시오. 예)팥 슈크림 초코:').split()

붕어빵_개수 = []
for i in 붕어빵_선택:
    choice_num = input(f'구매할 {i}붕어빵 개수를 입력하시오:')
    붕어빵_개수.append(choice_num)

주문_현황 = dict(zip(붕어빵_선택, 붕어빵_개수))


for a, b in 주문_현황.items():
    print(f'{a}붕어빵: {b}개')
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python

for a, b in 주문_현황.items():
    if int(붕어빵_재고[a]) - int(b) >= 0:
        붕어빵_재고[a] = int(붕어빵_재고[a])-int(b)
        print(f'{a}붕어빵: {b}개 판매')
    else:
        print(f'{a}붕어빵은 재고가 없습니다.')


```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
붕어빵_재고 = dict(팥=10, 슈크림=8, 초코=5)
붕어빵_관리 = 붕어빵_재고.items()

while True :
    print(f'팥붕아빵{붕어빵_재고["팥"]}개, 슈크림붕어빵{붕어빵_재고["슈크림"]}개, 초코붕아빵{붕어빵_재고["초코"]}개 남았습니다.')
    붕어빵_선택 = input('붕어빵 종류를 선택하시오. 예)팥 슈크림 초코:').split()
    if 붕어빵_선택 == ['종료']:
        break
    else:
        붕어빵_개수 = []
        for i in 붕어빵_선택:
            choice_num = input(f'구매할 {i}붕어빵 개수를 입력하시오:')
            붕어빵_개수.append(choice_num)

        주문_현황 = dict(zip(붕어빵_선택, 붕어빵_개수))
        for a, b in 주문_현황.items():
            if int(붕어빵_재고[a]) - int(b) >= 0:
                붕어빵_재고[a] = int(붕어빵_재고[a])-int(b)
                print(f'{a}붕어빵: {b}개 판매')
            else:
                print(f'{a}붕어빵은 재고가 없습니다.')
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
붕어빵_재고 = dict(팥=10, 슈크림=8, 초코=5)
붕어빵_관리 = 붕어빵_재고.items()

while True :
    print(f'팥붕아빵{붕어빵_재고["팥"]}개, 슈크림붕어빵{붕어빵_재고["슈크림"]}개, 초코붕아빵{붕어빵_재고["초코"]}개 남았습니다.')
    붕어빵_선택 = input('붕어빵 종류를 선택하시오. 예)팥 슈크림 초코:').split()
    if 붕어빵_선택 == ['종료']:
        break


    elif 붕어빵_선택 == ['1234']:
        a,b,c = map(int, input('추가할 붕어빵 갯수를 입력하시오.(팥, 슈크림, 초코 순 공백문자로 구분):').split())
        붕어빵_재고['팥']+=a; 붕어빵_재고['슈크림']+=b; 붕어빵_재고['초코']+=c
        continue

    else:
        붕어빵_개수 = []
        for i in 붕어빵_선택:
            choice_num = input(f'구매할 {i}붕어빵 개수를 입력하시오:')
            붕어빵_개수.append(choice_num)

        주문_현황 = dict(zip(붕어빵_선택, 붕어빵_개수))
        for a, b in 주문_현황.items():
            if int(붕어빵_재고[a]) - int(b) >= 0:
                붕어빵_재고[a] = int(붕어빵_재고[a])-int(b)
                print(f'{a}붕어빵: {b}개 판매')
            else:
                print(f'{a}붕어빵은 재고가 없습니다.')
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
붕어빵_재고 = dict(팥=10, 슈크림=8, 초코=5)
붕어빵_관리 = 붕어빵_재고.items()
붕어빵_가격 = dict(팥=1000, 슈크림=1200, 초코=1500)
판매_개수 = dict(팥=0, 슈크림=0, 초코=0)


while True :
    print(f'팥붕아빵{붕어빵_재고["팥"]}개, 슈크림붕어빵{붕어빵_재고["슈크림"]}개, 초코붕아빵{붕어빵_재고["초코"]}개 남았습니다.')
    붕어빵_선택 = input('붕어빵 종류를 선택하시오. 예)팥 슈크림 초코:').split()
    if 붕어빵_선택 == ['종료']:
        break


    elif 붕어빵_선택 == ['1234']:
        a,b,c = map(int, input('추가할 붕어빵 갯수를 입력하시오.(팥, 슈크림, 초코 순 공백문자로 구분):').split())
        붕어빵_재고['팥']+=a; 붕어빵_재고['슈크림']+=b; 붕어빵_재고['초코']+=c
        continue

    else:
        붕어빵_개수 = []
        for i in 붕어빵_선택:
            choice_num = input(f'구매할 {i}붕어빵 개수를 입력하시오:')
            붕어빵_개수.append(choice_num)

        주문_현황 = dict(zip(붕어빵_선택, 붕어빵_개수))
        for a, b in 주문_현황.items():
            if int(붕어빵_재고[a]) - int(b) >= 0:
                붕어빵_재고[a] = int(붕어빵_재고[a])-int(b)
                판매_개수[a] += int(b)
                print(f'{a}붕어빵: {b}개 판매')
            else:
                print(f'{a}붕어빵은 재고가 없습니다.')

총매출 = sum(판매_개수[i] * 붕어빵_가격[i] for i in 판매_개수)
print(f'총 매출: {총매출}원')
print('판매 개수:', 판매_개수)
```
